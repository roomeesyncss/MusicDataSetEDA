```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df=pd.read_csv('Taylor Swift Keys.csv')
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Pair plot
sns.pairplot(df, diag_kind='kde')
plt.suptitle('Pair Plot', y=1.02)
plt


```

```python
# Distribution of Danceability
plt.figure(figsize=(8, 6))
sns.histplot(df['Danceability'], bins=20, kde=True)
plt.title('Distribution of Danceability')
plt.xlabel('Danceability')
plt.ylabel('Frequency')
plt
```

```python
# Scatter plot of Danceability vs. Energy
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Danceability', y='Energy', data=df)
plt.title('Scatter Plot of Danceability vs. Energy')
plt.show()
```

```python
# Box plot of Danceability by Album
plt.figure(figsize=(12, 8))
sns.boxplot(x='Album', y='Danceability', data=df)
plt.title('Box Plot of Danceability by Album')
plt.xticks(rotation=45)
plt.show()
```

```python

# Bar plot of Average Danceability by Album
average_danceability_by_album = df.groupby('Album')['Danceability'].mean().sort_values(ascending=False)
plt.figure(figsize=(12, 8))
sns.barplot(x=average_danceability_by_album.index, y=average_danceability_by_album.values)
plt.title('Average Danceability by Album')
plt.xlabel('Album')
plt.ylabel('Average Danceability')
plt.xticks(rotation=45)
plt.show()
```

```python
# Swarm plot of Speechiness by Time Signature
plt.figure(figsize=(12, 8))
sns.swarmplot(x='Time Signature', y='Speechiness', data=df, palette='Set2')
plt.title('Swarm Plot of Speechiness by Time Signature')
plt.xlabel('Time Signature')
plt.ylabel('Speechiness')
plt.show()
```
